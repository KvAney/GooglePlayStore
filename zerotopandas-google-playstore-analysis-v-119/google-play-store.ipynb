So first import some python libraries to read and process our data

````python
import numpy as np
import pandas as pd
````

````python
df1 = pd.read_csv(data_dir +'/googleplaystore.csv')
````

 We are using the pandas library to read the dataset which is in csv form using function `.read_csv(<path_of_csv_file>)`

As we can see there are 13 columns . Most of them are of object type , which means pandas is not able to indetify appropriate datatype for the columns. Some columns such as Rating , Type , Content Rating , Current Ver , Android Ver have some NULL data  , i.e , missing data which is needed to be handled.


As only Rating has proper datatype form , only this column can be used to describe the dataset.
Rating is supposed to be in '1 - 5' margin.


Now we have to perform proper data cleaning.

One of the methods include handling missing data , i.e , properly handling missing values present in the dataset.

Now we can drop a few colums which are not usefull to us by using `.drop` 

````python
df = df1.drop(['Current Ver','Android Ver','Content Rating','Last Updated','Genres'],axis=1)
````

`.drop` function is used to drop/delete specified labels from rows or columns.

````python
df
````

````python
df.isna().sum()
````

Now Wherever are the rows that have nan values , it shall be replaced by the mean of that column.

````python
df['Rating'] = df['Rating'].fillna(df['Rating'].mean())
````

````python
df.isna().sum()
````

````python
df[df['Rating']> 5 ]
````

This is abnormal rating as it lies only between '1-5'

````python
df2= df.drop(10472)

df2.Type.fillna(value = 'Free'  , inplace = True)
````

````python
df2
````

Converting all the object type variable to a legitimate datatype

````python
df2['Price'] = df2['Price'].str.replace('$','')
df2['Installs'] = df2['Installs'].str.replace('+','')
df2['Installs'] = df2['Installs'].str.replace(',','')
df2['Size'] = df2['Size'].str.replace('M','')
df2['Size'] = df2['Size'].str.replace('k','')
df2['Size'] = df2['Size'].str.replace('+','')


df2
````

Now , `Price` , `Installs`  , `Reviews` are in *String* format.
We have to convert it into *int* or *float*

````python
df2['Reviews']=df2['Reviews'].apply(lambda x: int(x))
df2['Price']=df2['Price'].apply(lambda x: float(x))
df2['Installs']=df2['Installs'].apply(lambda x: int(x))
````

Hence , df2 is purest dataset uptil now.

````python
group = df2.groupby('Category')
RateRatio= group['Rating'].agg(np.mean)


plt.figure(figsize = (8,7))
plt.plot(RateRatio,'bo')
plt.xticks(rotation = 90)
plt.title('Category wise Rating')
plt.xlabel('Category')
plt.ylabel('Rating')
plt.show()
````

## Q1:
### Do the users like using most of the apps?

Let’s have a look at the distribution of the ratings of the data frame.

````python
plt.figure(figsize=(15,9))
plt.xlabel("Rating")
plt.ylabel("Frequency")
graph = sns.kdeplot(df2.Rating, color="Blue", shade = True)
plt.title('Distribution of Rating',size = 20);
````

We can clearly see that not many apps are having 1  or 2 rating . WHich means that most of the apps are good and liked by the audience

#### Q2: 
### Which category is most enjoyed by the users ?


````python
highest_Installs_df = df2.groupby('Category')[['Installs']].sum().sort_values(by='Installs', ascending=False)
````

`highest_Installs_df` is the dataframe that stores count of install per category

````python
highest_Installs_df.head()
````

````python
x2sis = []
y2sis = []

for i in range(len(highest_Installs_df)):
    x2sis.append(highest_Installs_df.Installs[i])
    y2sis.append(highest_Installs_df.index[i])

plt.figure(figsize=(18,13))

plt.xlabel("Installs")
plt.ylabel("Category")
graph = sns.barplot(x = x2sis, y = y2sis, alpha =0.9, palette= "viridis")
graph.set_title("Installs", fontsize = 25);
````

## Q3:
### Which app is the most expensive game available in the app store 

````python
def findtop10incategory(str):
    str = str.upper()
    top10 = df2[df2['Category'] == str]
    top10apps = top10.sort_values(by='Installs', ascending=False).head(10)
    # Top_Apps_in_art_and_design
    plt.figure(figsize=(15,12))
    plt.title('Top 10 Installed Apps',size = 20);    
    graph = sns.barplot(x = top10apps.App, y = top10apps.Installs)
    graph.set_xticklabels(graph.get_xticklabels(), rotation= 45, horizontalalignment='right');
````

````python
top10PaidApps_df = df2[df2['Type'] == 'Paid'].sort_values(by='Price', ascending=False).head(11)
# top10PaidApps
````

````python
top10PaidApps_df = top10PaidApps_df[['App', 'Installs']].drop(9934)
````

````python
plt.figure(figsize=(15,12));
plt.pie(top10PaidApps_df.Installs, explode=None, labels=top10PaidApps_df.App, autopct='%1.1f%%', startangle=0);
plt.title('Top Expensive Apps Distribution',size = 20);
plt.legend(top10PaidApps_df.App, 
           loc="lower right",
           title="Apps",
           fontsize = "xx-small"
          );
````

## Q5: 
### Which app is the most earning app in the google play store ?

````python
Paid_Apps_df = df2[df2['Type'] == 'Paid']
````

````python
earning_df = Paid_Apps_df[['App', 'Installs', 'Price']]
````

We can now add a separate column Earnings to our new data frame which we will create by multiplying the two-column *Price* and *Installs* .

````python
earning_df['Earnings'] = earning_df['Installs'] * earning_df['Price'];
````

````python
earning_df_sorted_by_Earnings = earning_df.sort_values(by='Earnings', ascending=False).head(50)
````

Now let us sort the above data by *Earnings* and *Price* .

````python
earning_df_sorted_by_Price = earning_df_sorted_by_Earnings.sort_values(by='Price', ascending=False)
````

We can plot the graph and find out which are the apps with the highest earning app.


````python
# PLot a bar chart of earning at y and app names at x
plt.figure(figsize=(15,9))
plt.bar(earning_df_sorted_by_Price.App, earning_df_sorted_by_Price.Earnings, width=1.1, label=earning_df_sorted_by_Price.Earnings)
plt.xlabel("Apps")
plt.ylabel("Earnings")
plt.tick_params(rotation=90)
plt.title("Top Earning Apps");
````
